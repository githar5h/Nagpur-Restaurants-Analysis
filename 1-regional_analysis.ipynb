```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

### Analysis 1: Regional Analysis


##### The distribution of restaurants across different areas and localities is explored and variance in costs in specific regions in analyzed.

```python
# Distribution of restaurants across areas
plt.figure(figsize=(20, 6))
sns.countplot(x='Area', data=df, order=df['Area'].value_counts().index)
plt.title('Distribution of Restaurants Across Areas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average cost for each area
average_cost_by_area = df.groupby('Area')['Cost for two'].mean().reset_index()

# Sort areas by average cost in descending order
sorted_areas = average_cost_by_area.sort_values(by='Cost for two', ascending=False)['Area']

# Bar plot of average cost across areas
plt.figure(figsize=(40, 6))
sns.barplot(x='Area', y='Cost for two', data=df, order=sorted_areas)
plt.title('Average Cost Across Areas (Descending Order)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average rating for each area
avg_rating_area = df.groupby('Area')['Rating'].mean().reset_index()

# Sort areas by average rating in descending order
sorted_areas = avg_rating_area.sort_values(by='Rating', ascending=False)['Area']

# Bar plot of average rating across areas
plt.figure(figsize=(40, 6))
sns.barplot(x='Area', y='Rating', data=avg_rating_area, order=sorted_areas)
plt.title('Average Rating Across Areas (Descending Order)')
plt.xlabel('Restaurant Area')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```
