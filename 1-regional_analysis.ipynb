```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

### Analysis 1: Regional Analysis


##### The distribution of restaurants across different areas and localities is explored and variance in costs in specific regions in analyzed.

```python
# Distribution of restaurants across areas
plt.figure(figsize=(20, 6))
sns.countplot(x='Area', data=df, order=df['Area'].value_counts().index)
plt.title('Distribution of Restaurants Across Areas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation between area and average cost
plt.figure(figsize=(40, 6))
sns.barplot(x='Area', y='Cost for two', data=df)
plt.title('Average Cost Across Areas')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze relation between area and Average Rating
avg_rating_area = df.groupby('Area')['Rating'].mean()
```

```python
# Visualize the relation between area and Average Rating using a bar plot
plt.figure(figsize=(40, 6))
sns.barplot(x=avg_rating_area.index, y=avg_rating_area.values)
plt.title('Average Rating Across Areas')
plt.xlabel('Restaurant Area')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```
