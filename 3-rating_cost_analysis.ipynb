```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

## Analysis 3: Rating and Cost Analysis

##### The relationship between restaurant ratings and the cost for two is examined, and the distribution of costs based on rating ranges is analyzed.

```python
# Create rating ranges
df['Rating Range'] = pd.cut(df['Rating'], bins=[0, 3, 4, 5], labels=['Low', 'Medium', 'High'])
```

```python
# Line plot of average cost for two across different rating ranges
plt.figure(figsize=(10, 6))
sns.lineplot(x='Rating Range', y='Cost for two', data=df, errorbar=None)
plt.title('Average Cost for Two Across Ratings')
plt.show()
```
