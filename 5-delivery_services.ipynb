```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

## Analysis 5: Delivery Services

##### The availability of long-distance delivery options is investigated and the correlation (if any) between the availability of long-distance delivery and higher ratings is analyzed.

```python
# Calculate average rating for restaurants with and without long-distance delivery
avg_rating_delivery = df.groupby('Long Distance Delivery')['Rating'].mean()

# Plotting the bar plot
avg_rating_delivery.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Rating by Long-Distance Delivery')
plt.xlabel('Long-Distance Delivery')
plt.ylabel('Average Rating')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```

```python
# Calculate average cost for restaurants with and without long-distance delivery
avg_cost_delivery = df.groupby('Long Distance Delivery')['Cost for two'].mean()

# Visualize the impact on cost using a bar plot
plt.figure(figsize=(8, 6))
avg_cost_delivery.plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Cost by Long-Distance Delivery')
plt.xlabel('Long-Distance Delivery')
plt.ylabel('Average Cost for Two')
plt.xticks([0, 1], ['No', 'Yes'], rotation=0)
plt.show()
```
