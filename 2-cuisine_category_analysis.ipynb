```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

## Analysis 2: Cuisine and Category Insights

##### The popularity of restaurant categories and cuisines is investigated to see how ratings differ among them.

```python
# Distribution of restaurant categories
plt.figure(figsize=(12, 6))
sns.countplot(x='Most Popular Category', data=df, order=df['Most Popular Category'].value_counts().index)
plt.title('Distribution of Restaurant Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average ratings by category
avg_ratings = df.groupby('Most Popular Category')['Rating'].mean().sort_values(ascending=False).index

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Most Popular Category', y='Rating', data=df, order=avg_ratings)
plt.title('Average Rating by Restaurant Category (Descending Order)')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate average cost for two by category
avg_cost_category = df.groupby('Most Popular Category')['Cost for two'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_cost_category.index, y=avg_cost_category.values)
plt.title('Average Price by Restaurant Category (Descending Order)')
plt.xlabel('Restaurant Category')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=90)
plt.show()
```
