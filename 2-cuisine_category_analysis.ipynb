```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

## Analysis 2: Cuisine and Category Insights

##### The popularity of restaurant categories and cuisines is investigated to see how ratings differ among them.

```python
# Distribution of restaurant categories
plt.figure(figsize=(12, 6))
sns.countplot(x='Most Popular Category', data=df, order=df['Most Popular Category'].value_counts().index)
plt.title('Distribution of Restaurant Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Average rating by category
plt.figure(figsize=(12, 6))
sns.barplot(x='Most Popular Category', y='Rating', data=df)
plt.title('Average Rating by Restaurant Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze Impact on Average Price
avg_price_category = df.groupby('Most Popular Category')['Cost for two'].mean()
```

```python
# Visualize the results using a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_category.index, y=avg_price_category.values)
plt.title('Average Price by Restaurant Category')
plt.xlabel('Restaurant Category')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=90)
plt.show()
```
