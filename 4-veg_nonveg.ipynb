```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset_cleaned.csv')
df.head()
```

## Analysis 4: Vegetarian vs. Non-vegetarian Preferences

##### The distribution and preferences of pure vegetarian and non-vegetarian restaurants in the dataset is explored.

```python
# Distribution of pure veg and non-veg restaurants
plt.figure(figsize=(8, 8))
df['Veg'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Restaurant is Pure Veg')
plt.show()
```

```python
# Swarm plot for ratings and pure veg status
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.swarmplot(x='Veg', y='Rating', data=df, size=4)
plt.title('Distribution of Ratings based on Pure Veg Status')
plt.xlabel('Pure Veg (True) vs. Non-Veg (False)')
plt.ylabel('Rating')
```

```python
# Strip plot for prices and pure veg status
plt.subplot(1, 2, 2)
sns.stripplot(x='Veg', y='Cost for two', data=df, jitter=True)
plt.title('Distribution of Prices based on Pure Veg Status')
plt.xlabel('Pure Veg (True) vs. Non-Veg (False)')
plt.ylabel('Cost for Two')
plt.show()
```
